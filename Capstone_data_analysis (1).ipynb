## Founder analysis

Next, we perform some analysis on the founder data included within the dataset, specifically how the number of founders relates to a company's growth

```python
# People strategy - how the number of founders relates to a company's growth

# growth_cols = ['Number of Employees', 'Number of Founders', 'Last Funding Amount', 'Estimated Revenue Range', 'Number of Funding Rounds']

growth_cols = ['Number of Employees', 'Last Funding Amount (in USD)', 'Number of Funding Rounds']
selected_cols = ['Number of Founders'] + growth_cols
selected_df = data2[selected_cols]

# Plotting scatter plots
for col in growth_cols:
    plt.figure(figsize=(8, 6))
    plt.scatter(selected_df['Number of Founders'], selected_df[col], alpha=0.5)
    plt.xlabel('Number of Founders')
    plt.ylabel(col)
    plt.title('Number of Founders vs {}'.format(col))
    plt.grid(True)
    plt.show()
```

```python
# Function that makes employee ranges easier to handle by giving them a numerical average value

def convert_employee_range(employee_range):
    if isinstance(employee_range, str):
        parts = employee_range.split('-')
        if len(parts) == 2:
            return (int(parts[0]) + int(parts[1])) / 2
        elif '1001-5000' in employee_range:
            return 3000
        elif '5001-10000' in employee_range:
            return 7500
        elif '10001+' in employee_range:
            return 10001  
    return None

data2['Number of Employees New'] = data2['Number of Employees'].apply(convert_employee_range)
data2
```

The following scatter plots show how some key variables of interest change as the number of founders changes. 

The key interesting insights include: 

Funding amount decreases as the number of founders in a company increases, but more significantly from 4 founders onwards

The same applies for number of employees i.e. fewer companies with more than 4 founders have companies with a headcount of more than 150 employees

## Revenue analysis

```python
# Box plot: Number of Founders vs Estimated Revenue Range - really useful plot!!
plt.figure(figsize=(12, 8))
sns.boxplot(x='Number of Founders', y='Estimated Revenue Range', data=data2, order=data2['Estimated Revenue Range'].value_counts().index)
plt.title('Number of Founders vs Estimated Revenue Range')
plt.xlabel('Number of Founders')
plt.ylabel('Estimated Revenue Range')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Adjusting the approach for the bar plot to handle categorical data better
# Create a mapping for the estimated revenue range to numeric values for better plotting
revenue_mapping = {
    "<$1M": 0.5,
    "$1M to $10M": 5,
    "$10M to $50M": 30,
    "$50M to $100M": 75,
    "$100M to $500M": 300,
    "$500M to $1B": 750,
    "$1B to $10B": 5000,
    "$10B+": 10000
}

data2['Revenue Range Numeric'] = data2['Estimated Revenue Range'].map(revenue_mapping)

data2['Revenue Range Numeric'] = pd.to_numeric(data2['Revenue Range Numeric'], errors='coerce')

# Calculate the median revenue range per number of founders
median_revenue_by_founders = data2.groupby('Number of Founders')['Revenue Range Numeric'].median().dropna()

plt.figure(figsize=(10, 6))
median_revenue_by_founders.plot(kind='bar')
plt.title('Median Estimated Revenue Range by Number of Founders')
plt.xlabel('Number of Founders')
plt.ylabel('Median Estimated Revenue Range (in millions)')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

## Investors

This investor analysis focuses on two main areas: 

Identifying Common Investors: Extract and count the frequency of each investor across all companies. <br>
Analysing Impact on Growth Metrics: Compare growth metrics (funding amount, funding rounds, employee numbers, estimated revenue) for companies with common investors.

```python
from collections import Counter

# Split the 'Top 5 Investors' column into individual investors
investors_list = data2['Top 5 Investors'].dropna().str.split(',').tolist()
flattened_investors_list = [investor.strip() for sublist in investors_list for investor in sublist]

# Counting the frequency
investor_counts = Counter(flattened_investors_list)
```

```python
# Converting to DF
investor_counts_df = pd.DataFrame.from_dict(investor_counts, orient='index', columns=['Frequency']).reset_index()
investor_counts_df.rename(columns={'index': 'Investor'}, inplace=True)

# Top investors
top_investors = investor_counts_df.sort_values(by='Frequency', ascending=False).head(10)
print(top_investors)
```

```python
top_investors_list = top_investors['Investor'].tolist()

# Function that filters companies that have these top investors
def has_top_investor(investors):
    if pd.isna(investors):
        return False
    return any(investor in investors for investor in top_investors_list)

# New column for top investors
data2['Has Top Investor'] = data2['Top 5 Investors'].apply(has_top_investor)

# Growth metrics
growth_metrics = ['Last Funding Amount (in USD)', 'Number of Funding Rounds', 'Number of Employees New', 'Revenue Range Numeric']
median_growth_metrics = data2.groupby('Has Top Investor')[growth_metrics].median().reset_index()

print(median_growth_metrics)
```

Perhaps the most interesting result from this analysis of top investors was the presence of top investors incrasing funding rounds, potentially leading to higher chances of scaling up. However, the number of top investors isn't large enough to allow any significant conclusions to be drawn. 

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Last Funding Amount (in USD)
sns.boxplot(x='Has Top Investor', y='Last Funding Amount (in USD)', data=data2, ax=axes[0, 0])
axes[0, 0].set_title('Impact of Top Investors on Last Funding Amount (in USD)')

# Number of Funding Rounds
sns.boxplot(x='Has Top Investor', y='Number of Funding Rounds', data=data2, ax=axes[0, 1])
axes[0, 1].set_title('Impact of Top Investors on Number of Funding Rounds')

# Number of Employees
sns.boxplot(x='Has Top Investor', y='Number of Employees New', data=data2, ax=axes[1, 0])
axes[1, 0].set_title('Impact of Top Investors on Number of Employees')

# Revenue Range Numeric
sns.boxplot(x='Has Top Investor', y='Revenue Range Numeric', data=data2, ax=axes[1, 1])
axes[1, 1].set_title('Impact of Top Investors on Revenue Range Numeric')

plt.tight_layout()
plt.show()
```

```python
# Using only top investors (based on their frequency of investment)
import networkx as nx

top_investors_list = top_investors['Investor'].tolist()

G = nx.Graph()
# Link between investors and companies
for index, row in data2.iterrows():
    if pd.notna(row['Top 5 Investors']):
        investors = row['Top 5 Investors'].split(',')
        for investor in investors:
            investor = investor.strip()
            if investor in top_investors_list:
                G.add_edge(row['Organization Name'], investor)

# Displaying the graph
plt.figure(figsize=(25, 20))
pos = nx.spring_layout(G, k=0.5)
nx.draw(G, pos, with_labels=True, node_size=100, node_color='pink', font_size=14, font_color='black', edge_color='grey')
plt.title('Investor-Company Network (Top Investors)')
plt.show()
```

## Funding stage

This will help quantify funding stage, which will be useful for our ML model. We categorised Pre-Seed, Seed, Series A as early stage, while all other stages were late-stage. 

```python
data2['Last Funding Type'].unique()
```

```python
early_stage = ['Pre-Seed', 'Seed', 'Series A']
late_stage = ['Series B', 'Debt Financing', 'Non-equity Assistance', 'Venture - Series Unknown', 
              'Private Equity', 'Grant', 'Equity Crowdfunding', 'Post-IPO Equity', 'Convertible Note', 
              'Undisclosed', 'Post-IPO Debt', 'Corporate Round', 'Secondary Market', 
              'Product Crowdfunding']

# Create a column for Funding Stage
data2['Funding Stage'] = data2['Last Funding Type'].apply(lambda x: 'Early' if x in early_stage else ('Late' if x in late_stage else 'Other'))

# Compare growth metrics for early vs. late-stage investments
stage_metrics = data2.groupby('Funding Stage')[['Last Funding Amount (in USD)', 'Number of Funding Rounds', 'Revenue Range Numeric']].median()

print(stage_metrics)
```

Higher last funding amount, number of funding rounds and revenue range observed for early vs late stage. 

```python
stage_metrics2 = data2.groupby('Funding Stage')[['Last Funding Amount (in USD)', 'Number of Funding Rounds', 'Revenue Range Numeric']].mean()

print(stage_metrics2)
```

## Investment Trends

```python
# Count of investments

data2['Founded Date DT'] = pd.to_datetime(data2['Founded Date'], errors='coerce')

# Group by year and calculate the number of investments
investment_trends = data2.groupby(data2['Founded Date DT'].dt.year)['Last Funding Amount (in USD)'].count()

# Filter the data to include only from 2000 onwards
investment_trends = investment_trends[investment_trends.index >= 2000]

plt.figure(figsize=(10, 6))
investment_trends.plot(kind='bar', color='brown', edgecolor='black')
plt.title('Investment Trends Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Investments', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```

```python
# Sum of investments

data2['Founded Date DT'] = pd.to_datetime(data2['Founded Date'], errors='coerce')

# Group by year and calculate the sum of investments
investment_trends = data2.groupby(data2['Founded Date DT'].dt.year)['Last Funding Amount (in USD)'].sum()

# Filtering data to only include 2000 onwards
investment_trends = investment_trends[investment_trends.index >= 2000]

plt.figure(figsize=(10, 6))
investment_trends.plot(kind='bar', color='brown', edgecolor='black')
plt.title('Investment Trends Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Sum of Investments', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```
