## Demographic Analysis

```python
# Reducing duplicates

data['Location'] = data['Location'].str.strip().str.title()
data['Primary Job Title'] = data['Primary Job Title'].str.strip().str.title()
data['Schools Attended'] = data['Schools Attended'].str.strip().str.title()
data['Current Organizations'] = data['Current Organizations'].str.strip().str.title()
```

```python
# Gender breakdown
gender_distribution = data['Gender'].value_counts()

plt.figure(figsize=(8, 6))
gender_distribution.plot(kind='bar', color=['blue', 'pink', 'green'])
plt.title('Gender Distribution of Founders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

gender_distribution
```

```python
location_distribution = data['Location'].value_counts()
location_distribution
```

```python
# Creating a function for the above analysis 

def demographic_analysis(data, column_name):
    """
    Perform demographic analysis on a specified column in the dataframe, using the two parameters data and column_name.
    
    Returns:
        pd.Series: A series containing the top 10 most frequent values in the specified column.
    """
    
    # Getting the top 10 values
    top_10_distribution = data[column_name].value_counts().head(10)
    
    # Plot the distribution
    plt.figure(figsize=(8, 6))
    top_10_distribution.plot(kind='bar', color='skyblue')
    plt.title(f'Top 10 Most Frequent Values in {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.grid(axis='y')
    plt.show()
    
    return top_10_distribution
```

```python
# Testing the function with the 'Location' column
location_distribution = demographic_analysis(data, 'Location')
location_distribution
```

```python
org_distribution = demographic_analysis(data, 'Primary Organization')
org_distribution
```

```python
school_distribution = demographic_analysis(data, 'Schools Attended')
school_distribution
```

```python
title_distribution = demographic_analysis(data, 'Primary Job Title')
title_distribution
```

Analysing if co-founding teams are more successful than solo founders based on the number of exits (IPO)

```python
# CoFounder vs Founder comparison

# Creating a new column 'Team Type'

def classify_team_type(title):
    if isinstance(title, str):
        title_lower = title.lower()
        if 'co-founder' in title_lower:
            return 'Co-Founder'
        elif 'founder' in title_lower:
            return 'Founder'
    return 'Other'

# Applying function

data['Team Type'] = data['Primary Job Title'].apply(classify_team_type)
data
```

```python
data['Team Type'].value_counts()
```

```python
# Converting 'Number of Exits (IPO)' to numeric
data['Number of Exits (IPO)'] = pd.to_numeric(data['Number of Exits (IPO)'], errors='coerce')

# Calculating mean
success_by_team_type = data.groupby('Team Type')['Number of Exits (IPO)'].mean()

# Plotting results

plt.figure(figsize=(10, 6))
success_by_team_type.plot(kind='bar', color=['blue', 'orange', 'red'])
plt.title('Average Number of Exits (IPO) by Team Type')
plt.xlabel('Team Type')
plt.ylabel('Average Number of Exits (IPO)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

success_by_team_type
```

```python
# T-test to check for statistical significance 
from scipy.stats import ttest_ind

co_founders = data[data['Team Type'] == 'Co-Founder']['Number of Exits (IPO)'].dropna()
solo_founders = data[data['Team Type'] == 'Founder']['Number of Exits (IPO)'].dropna()

# Perform the t-test
p_value = ttest_ind(co_founders, solo_founders)

p_value
```
